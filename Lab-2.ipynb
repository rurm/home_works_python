#### Задание 4.
Визуализировать распределение переменной ‘Price’, рассчитать показатели skewness и kurtosis

```python
import matplotlib.pyplot as plt
price0=list(auto_data.price)
plt.hist(price0,bins=10)
# похоже на логнормальное распределение, проверим - прологорифмиуем price0 и построим гистограмму
```

```python
from math import log
plt.hist([log(y,10) for y in price0],bins=10)
# теперь больше похоже на нормальное, по крайне мере больше чем предыдущее
```

```python
from scipy.stats import skew
skew(auto_data.price)
```

```python
from scipy.stats import kurtosis
kurtosis(auto_data.price)
```
